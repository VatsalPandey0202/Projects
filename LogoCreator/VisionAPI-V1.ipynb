```python
import io
import os
import cv2
import glob

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] ="C:/Users/Vatsal/OneDrive/Desktop/ace-agility-315321-ce7731113c3f.json"
imdir = 'D:/SVG_LogoGenerator-ImageGAN/Data/Conditional_StyleGAN_Logo/'
ext = ['png', 'jpg', 'gif']    # Add image formats here
files = []
[files.extend(glob.glob(imdir + '*.' + e)) for e in ext]
name=[]
piclabels=[]
c=0
# Imports the Google Cloud client library
from google.cloud import vision

for f in files:
    name.append(f[61:])
    #print(f[61:])
    # Instantiates a client
    client = vision.ImageAnnotatorClient()

    # Loads the image into memory
    with io.open(f, 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

    # Performs label detection on the image file
    response = client.label_detection(image=image)
    labels = response.label_annotations
    temp=[]
    for label in labels:
        temp.append(label.description)
    piclabels.append(temp)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(name, piclabels)),
               columns =['Name', 'Labels'])
df
```
